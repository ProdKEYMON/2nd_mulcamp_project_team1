```python
# Document Loaders
from langchain.document_loaders import WebBaseLoader

web_site = 'https://ko.wikipedia.org/wiki/%EB%89%B4_%EC%9E%AD_%EC%8A%A4%EC%9C%99'
# 웹페이지 불러오기
loader = WebBaseLoader(web_path = web_site)
documents = loader.load()
# documents
```

```python
# TextSplitter
# 웹페이지 내용이 많기 때문에 chatgpt가 이해하기 쉽게 텍스트 분할
# (현실적으로) 비용 아껴야함

from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 0)
docs = text_splitter.split_documents(documents)
```

```python
# 요약하기
# 필요한 라이브러리 임포트
from langchain.document_loaders import WebBaseLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.chains.summarize import load_summarize_chain
from langchain.chains import LLMChain
from langchain.chat_models import ChatOpenAI
from langchain.prompts.chat import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate
)

chat = ChatOpenAI(temperature = 0)
summary_chain = load_summarize_chain(chat, chain_type='map_reduce', verbose=True)
summary = summary_chain.run(docs)

# 번역 파트

template = 'You are a helpful assistant that translates {input_language} to {output_language}. Text: {text}'
system_message_prompt = SystemMessagePromptTemplate.from_template(template)
human_message_prompt = HumanMessagePromptTemplate.from_template(template)
messages = [system_message_prompt, human_message_prompt]
chat_prompt = ChatPromptTemplate.from_messages(messages)

chatchain = LLMChain(llm=chat, prompt=chat_prompt)
translated_text = chatchain.run(input_language='English', output_language='Korean', text=summary)

print("번역 결과: ", translated_text)
```
